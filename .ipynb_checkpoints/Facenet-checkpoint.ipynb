AUTOMATED PREPROCESSING


```python
import os
import cv2 as cv
import numpy as np
from mtcnn import MTCNN
import matplotlib.pyplot as plt

class FaceRender:
    def __init__(self, directory):
        self.directory = directory
        self.target_size = (160, 160)
        self.X = []
        self.Y = []
        self.detector = MTCNN()

    def extract_face(self, file_path):
        img = cv.imread(file_path)
        if img is None:
            return None
        img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        faces = self.detector.detect_faces(img)
        if len(faces) == 0:
            return None
        x, y, w, h = faces[0]['box']
        x, y = abs(x), abs(y)
        face = img[y:y+h, x:x+w]
        face_resized = cv.resize(face, self.target_size)
        return face_resized

    def load_faces(self, dir_path):
        faces = []
        for im_name in os.listdir(dir_path):
            path = os.path.join(dir_path, im_name)
            # Skip non-image files
            if not im_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                continue
            try:
                face = self.extract_face(path)
                if face is not None:
                    faces.append(face)
            except Exception as e:
                print(f"Failed to load {im_name}: {e}")
                continue
        return faces

    def main(self):
        for sub_dir in os.listdir(self.directory):
            path = os.path.join(self.directory, sub_dir)
            #  Skip files and unwanted directories
            if not os.path.isdir(path):
                continue
            if sub_dir.startswith('.') or 'ipynb_checkpoints' in sub_dir.lower():
                continue

            print(f"Loading from {path} ...")
            faces = self.load_faces(path)
            labels = [sub_dir for _ in range(len(faces))]
            print(f"Loaded successfully : {len(labels)}")
            self.X.extend(faces)
            self.Y.extend(labels)

        return np.asarray(self.X), np.asarray(self.Y)

    def plot_images(self):
        if len(self.X) == 0:
            print("No images to plot.")
            return
        ncols = 3
        nrows = len(self.X) // ncols + 1
        plt.figure(figsize=(10, 10))
        for i, img in enumerate(self.X):
            plt.subplot(nrows, ncols, i + 1)
            plt.imshow(img)
            plt.axis('off')
        plt.show()
```

```python
fr = FaceRender("C:/Users/Satyam/Desktop/Facenet/dataset/")
X, Y = fr.main()
```

```python
fr.plot_images()
```

```python
from keras_facenet import FaceNet
embedder = FaceNet()

def get_embedding(face_img):
    face_img =face_img.astype('float32')
    face_img = np.expand_dims(face_img,axis=0)
    yhat=embedder.embeddings(face_img)
    return yhat[0]
    
```

```python
EMBEDDED_X= []
for img in X:
    EMBEDDED_X.append(get_embedding(img))
EMBEDDED_X = np.asarray(EMBEDDED_X)
```
